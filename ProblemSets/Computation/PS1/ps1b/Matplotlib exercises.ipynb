```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Problem 1
def var_random_normal(n):
    x = np.random.normal(size=(n,n))
    y = np.mean(x,axis = 1)
    return np.var(y)

def array_vars_random_normal():
    n = np.arange(100,1100,100)
    return [var_random_normal(i) for i in n]

array_vars_random_normal()
```

```python
# Problem 2
def plot_fun(a,b):
    x = np.linspace(a,b,100)
    plt.plot(x,np.sin(x),label='sin')
    plt.plot(x,np.cos(x),label='cos')
    plt.plot(x,np.arctan(x),label='arctan')
    plt.legend()
    
plot_fun(-2*np.pi,2*np.pi)
```

```python
# Problem 4
x = np.linspace(0,2*np.pi,100)
ax1 = plt.subplot(2,2,1)
ax1.plot(x,np.sin(x),'g')
ax1.set_title('sin(x)')

ax2 = plt.subplot(2,2,2)
ax2.plot(x,np.sin(2*x),'r--')
ax2.set_title('sin(2x)')

ax3 = plt.subplot(2,2,3)
ax3.plot(x,2*np.sin(x),'b--')
ax3.set_title('2sin(x)')

ax4 = plt.subplot(2,2,4)
ax4.plot(x,2*np.sin(2*x),'m.')
ax4.set_title('2sin(2x)')

plt.suptitle('wa!')
```
